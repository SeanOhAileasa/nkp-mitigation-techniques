###### 2. Signature Management
***

After completing this topic, you should be able to: ``describe signature management``

Now, we'll begin this course with a look at what a virus signature is. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Virus Signature
***

And this is quite simply a string of characters or numbers that is assigned to any given virus as a means of identifying it so that antivirus programs and/or possibly intrusion detection systems can pick up on the presence of that threat because it recognizes the signature.

Now, signatures use algorithms or hashes to uniquely identify a specific virus. But in fact, multiple viruses can share a single signature. Because in many cases, there are, quite simply, different strains of the same virus. So by picking up on that single signature, it might actually be able to pick up on the varieties of strains that might be out there. Because they do share the same core, if you will. And that's what the antivirus program looks for. So it allows it to pick up on these variations. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Antivirus Vendors
***

So when it comes to the antivirus vendors, they issue the signatures and they release them on an ongoing basis and often within hours of new threats being discovered. And if you're curious as to how they are able to pick up on these threats so quickly, they quite literally implement data centers where people do little more then attempt to infect their systems intentionally. 

They are doing everything that they can on the Internet that would normally be considered risky or dangerous when it comes to getting a virus. 

So they do everything possible to try to contract them, and this helps them to keep on top of anything that is picked up. And, of course, they rely on the installations of their own software as well to report back if they discover something that does seem to be at least a little bit suspicious. They might not recognize any given virus by its signature because it could be brand new. But in many cases, they are still able to pick up on things that are suspicious, if you will. And in fact, that goes along with the next point here, that it's not always just signature management. 

Vendors are also offering solutions such as file reputation and heuristics-based engines.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### File Reputation
***

Now, file reputation, in essence, can track a file as it goes through changes, as it moves through your environment and, in essence, can learn what is normal with respect to the changes that might occur.

But any kind of infection is a change. So if a change happens that seems to be outside of the norm, then that is considered to be suspicious and it can be reported.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Heuristics-based Engines
***

Heuristics-based engines, basically, don't rely on signatures. They, quite literally, try to apply almost a human logic to determine whether or not there is anything wrong with this file. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Signatures
***

So signatures are by no means 100%. They're certainly very effective for known viruses, but of course, new viruses that just come out will not have a signature. 

So they need to be captured by some other means, and then they will likely be given a signature. But, of course, if has to be captured in the first place. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Management
***

Now with respect to management, you will almost always find in a corporate environment some kind of centralized antivirus management implementation.

This allows for very granular control over the client systems with respect to how and when they receive their updates, as well as ensuring that they are all configured consistently. 

An administrator can sit at a server, that is, the management server, and define policies. And this one policy, or a few different ones, can then be deployed to all of the client systems so that they have the same configuration. 

They are set to scan with the same settings. They are set to check for updates at the same time. So this way you know that all of your systems are protected and they have this consistent configuration.

You can also use those applications to monitor the users on a network. So you can find out if everyone has the same and the most up-to-date version. 

You can receive alerts when any kind of infection is detected or how widespread it is. 

And, of course, you can update the virus signatures from the central management server and then deploy those updates to the client.

So, again, this is something that you definitely see in most corporate environments. 

For home users, you generally just install the software. And, of course, you can configure it however you like. But it's still important to ensure that it is set to scan. And, of course, set to automatically go out and check for updates. 

So the administrators can push out new signature updates to the clients on a network, as mentioned. And in many cases, you have a point and click graphic user interface to do this from. But you might also have the ability to script it. And this might be an option depending on how similar all of your systems are. If you have quite a mix of different operating systems and different types of devices, you might need to script something just for compatibility reasons. 

But if everything is fairly consistent, you generally can just use a graphic interface and say, here are the updates. And again, you configure the clients with their policies to check for the updates on a regularly scheduled basis. 

But you can, in many cases, also force them to check right now. If, for example, you just received new updates and you know that their update cycle is maybe to only check every three days, and let's say you just got these updates today but you know they checked yesterday, then you know it's going to be two more days before they check again. So you can, in essence, force them to refresh and get the new updates.

<a id="___"></a>
***
###### Automated Update Services
***

Now the automated update services, in essence, are what the vendor themselves provide with respect to when they release the updates. 

So your management server can periodically check with the vendor for new signatures or any kind of program upgrade. 

So, again, it's up to you as an administrator to determine when your management server goes out to check for the updates. 

It usually gets them, pulls them down, then you distribute them to the clients. 

But it might also be a new version of the software itself to address any kind of stability issues, or compatibility issues, things along those lines. But, ultimately, when you do implement any kind of antivirus solution in your environment, this type of enterprise configuration allows much greater control and much greater consistency for the way the client systems are protected.

###### 4. VLAN
***

After completing this topic, you should be able to: ``recognize when to change native VLANs``

Now in this presentation, we'll take a look at some VLAN configuration options. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### VLAN Configuration Options
***

As, these days, many Layer 2 switches allow you to group the physical switch ports into logical virtual LANs, or VLANs. 

And each VLAN on a switch represents an individual broadcast domain.

So, for example, a broadcast on a switch port belonging to VLAN 1 will not be seen on a switch port belonging to VLAN 2. 

Now, also be aware that a VLAN can exist on more than one switch. 

In other words, you can go in both directions. If, for example, you have a single switch, and let's just say there were 24 ports, you could absolutely create two VLANs of 12 each within that single switch. Or you could go the other way. If you had two physical switches, each with 24 ports, you'd have 48 in total, you could create a single VLAN out of all 48. That's entirely up to you.

Now in most IP environments, each VLAN is usually associated with an IP subnet, but that's just common, it's not required. But once you do have your subnets defined and mapped to the VLANs, then traffic does need to be routed to get any traffic from one VLAN to another. Because, of course, once it is a separate IP subnet, to get to another IP subnet, it does have to be routed. But, again, that's just common. You could take a single subnet and divide it into multiple smaller VLANs. In other words, segment that subnet if you want to. Again, the configuration is up to you. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### VLAN Trunking Protocol - VTP
***

Now the VLAN trunking protocol, or VTP, allows VLAN changes on one switch to be dynamically advertised to adjacent switches via the trunk links. 

You can implement a VTP server which can be used to add, delete, or modify the VLAN information.

And that information is then propagated over the trunk links and learned by other VTP servers and VTP clients.

And this just allows you to centralize the management. So if you have a particularly dynamic environment, I can centrally control the configuration of all the VLANs from the VTP server. And this, of course, allows me to push out those changes, if you will, very easily. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Native versus Default VLANs
***

Now when it comes to the native versus the default VLAN, there's just a little bit of terminology, really, that needs to be clarified here. 

In most cases, you'll hear people refer to native and/or default as meaning the same thing. 

But a native VLAN, in essence, is defined by the 802.1Q specification, wherein one VLAN in the trunk is known as the native VLAN. 

And, as such, the native VLAN does not receive any tagging, but all other VLANs do receive a tag.

And the tag is embedded in one of the four extra bytes that are prepended to each Ethernet frame in a non-native VLAN. 

So the native VLAN, in essence, exists by default. 

So, again, they are very similar terms, but VLAN 1 is always the default VLAN and is mandatory. And this is needed for many protocol communications between switches, such as the Spanning-Tree Protocol, which we'll actually talk about in the next presentation.

But just coming back to that native versus default, again, the native VLAN is always there. But in terms of the behavior of the packets that are being switched, if there is a tag, then that indicates which VLAN the packet is destined for. 

If there is no tag, then it will, by default, be sent to the native VLAN. 

So that, again, is just simply a little bit of terminology. The VLAN itself is referred to as the native VLAN, but the switching behavior is for untagged packets to be delivered by default to the native VLAN. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Conclusion - Native versus Default VLANs
***

So, again, most people will probably use those terms somewhat interchangeably. But one refers to the VLAN itself. One refers to the switching behavior when it comes to those untagged packets. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Conclusion - VLAN Configuration
***

So, ultimately, it's up to you how you configure your VLANs. But, again, things like the native and the VTP server can certainly help you ease the configuration and determine where your VLANs are going to reside.

###### 5. Switch Port Configuration
***

After completing this topic, you should be able to: ``recognize switch port configuration using the Spanning Tree Protocol, BPDU guard, and root guard``

In this presentation we'll introduce a Switch Configuration Protocol known as STP, or the Spanning Tree Protocol. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Spanning Tree Protocol (STP)
***

Now this is something that is implemented when you have redundant switches. 

And that is often desirable because it gives you some fault tolerance. 

If one switch goes down, another one is available. And this creates an alternate path to get to the ultimate destination. 

But this redundancy can introduce an additional problem known as switching loops. And this is because switches learn where any given system is in terms of having to forward the packets. So as each switch discovers that any particular system is accessible over two different paths, it can end up simply cycling that packet around in an endless loop. And that can very quickly bring a network down. 

So the Spanning Tree Protocol allows you to implement a loop-free but yet still redundant topology by assigning roles to the switches. 

And the available roles are the root bridge, the root port, a designated port and a non-designated port. 

Now the switch itself is either a root bridge or a non-root bridge. 

And then on a root bridge you will always see designated ports specifying that this port is the path to this segment. [A diagram shows two segments, Segment 1 and Segment 2, connected by two switches, the Root bridge and the Non-root bridge. Designated ports on the Root bridge connect to each segment. On the Non-root bridge, a Root port connects to Segment 1 and a Non-designated port connects to Segment 2. ] 

A non-root bridge, in essence, does not make any decisions as to how traffic will be moved. It just has its own root port for that switch. 

But then we see a non-designated port which, in simple terms, is a port that's turned off. 

So the root port is on, the non-designated port is off. So since we see one of them closed, if you will, that prevents the looping from happening. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Spanning Tree Protocol (STP) - 802.1d Standard (Slow)
***

Now the original Spanning Tree Protocol was defined by 802.1d. 

And by default, it took 50 seconds for a port to fully transition from the blocking state, or that non-designated, to the forwarding state, or the root. 

And, of course, that was a significant delay. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Rapid Spanning Tree Protocol (RSTP) - 802.1w Standard
***

So a newer version of the spanning tree known as Rapid Spanning Tree Protocol, or RSTP, was introduced in 1998 And defined by the 802.1w standard. 

And RSTP can reconverge all of the states of the ports in approximately three seconds, so a significant improvement with RSTP. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Rapid Spanning Tree Protocol (RSTP) - Configuration
***

Now it implements this configuration by advertising whats known as Bridge Protocol Data Units or BPDUs. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### RSTP - Bridge Protocol Data Units (BPDUs)
***

When a switch first boots up it starts sending its own BPDUs. 

When it learns the identity of the root bridge, which is something you configure, only the root bridge continues to source out BPDUs. In other words, all of the other switches withdraw. 

The BPDUs are then forwarded throughout the network to control the state of any given port. And those four states are:


i. Blocking


ii. Listening


iii. Learning 


iv. Forwarding

So it has to transition through each of those before it will effectively become a forwarding port or one that handles traffic. 

So it is blocked by default. But if a switch realizes that another switch has gone down, then it can effectively transition so that the blocking port becomes available. And, of course, it can continue to provide services so that you don't lose the entire network with the loss of that switch. Because, again, that's why you implemented the redundancy in the first place.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Bridge Protocol Data Units (BPDU) Guard
***

Now there are also some guards in place with BPDUs. 

And one is specifically known as the BPDU Guard. 

Which prevents switching devices from connecting to any other switching device that has PortFast-enabled ports. 

A PortFast-enabled port is one that immediately sets itself to a forwarding state instead of going through those four state changes that we just saw. This reintroduces the possibility of the loop occurring. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Root Guard
***

And then root guard is then next one. 

And this, quite simply, prevents switches from becoming a root switch in the first place. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Loop Guard
***

And, finally, loop guard improves the stability in preventing bridging loops by adding another state, if you will, we just talked about those four states.

But if the non-designated port with this feature enabled stops receiving BPDUs it kind of assumes that there is a problem. 

For example, another switch has gone down. But instead of going through the four states at that point, it places itself into what's known as a loop inconsistent but still-blocking state.

Basically, to give the other switch time to come back. 

After a pre-configured amount of time, it will then go through the four states of blocking, listening, learning and forwarding before it actually implements itself as a forwarding port. 

So it just gets you that extra step to allow the original switch to come back online.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Conclusion - Spanning Tree Protocol - Redundancy for Fault Tolerance
***

But, ultimately, the Spanning Tree Protocol is for when you do want redundancy for fault tolerance purposes. But you still want to have a network that is, of course, not going to immediately go down because these loops are occurring.

###### 6. Switch Port Protection
***

After completing this topic, you should be able to: ``recognize switch port protection such as ARP poisoning, DHCP snooping, and MAC filtering``

In this presentation we'll take a look at some switch configuration options pertaining to security, beginning with DHCP snooping.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Dynamic Host Configuration Protocol (DHCP) Snooping
***

Almost every environment these days will implement DHCP in one form or another.

So DHCP snooping is used to enforce security in a DHCP network and is configured at the switch port itself. 

It uses DHCP information to determine whether or not any given DHCP server is acceptable. 

And if it's deemed to be unacceptable, it quite simply drops the traffic. And this ensures that only authorized DHCP servers are used. Or perhaps more to the point, it prevents unauthorized DHCP servers from offering IP addresses to the clients. 

So using a fairly simple example, let's assume you have a single DHCP server. And you have it plugged into physical switch port number one. As such, that switch will expect DHCP traffic to pass only over that port. So if someone attempts to create a rogue DHCP server, to hand out bogus IP addresses, it would, of course, have to be plugged into a different physical switch port. At which point the switch would say, wait a second, I'm not supposed to be seeing DHCP traffic over that port. Therefore, that server is unauthorized and it quite simply drops the traffic. 

Now, I should also mention that if you are in a Windows Active Directory environment and you are using a Windows server for DHCP. Then Active Directory itself can also authorize that DHCP server. And again, more to the point, your Active Directory clients will not listen to a DHCP server that has not been authorized. But that's much more of a logical configuration, again, only within Active Directory. 

There are many other servers, many other devices that can handle DHCP services. Routers, other operating systems can easily perform that task. So this is a much lower level configuration, if you will, right on the switch port. And, again, it simply would not care what you are using for your DHCP services. It is based entirely on the physical port itself. And only those configured to be acceptable will pass the DHCP traffic. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Address Resolution Protocol (ARP) Poisoning
***

ARP poisoning is another option you can configure. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Dynamic ARP Inspection
***

And this uses a protocol known as Dynamic ARP Inspection. 

This is also configured on the switch port. 

And again, just prevents ports from passing traffic where the IP address and the MAC address don't match what's in a trusted database of stored pairings. 

And this is something that's built at runtime in combination, in many cases, with DHCP snooping. 

Because once the DHCP server issues out an authorized IP address to a client, then it registers the fact that this MAC address is currently using this IP address. And basically that is entered into a trusted database. So that if somebody attempts to poison the cache by altering the values of the cache, the switch can say, nope, that does not match the pairing that I see in the trusted database. So ultimately this defends against ARP poisoning attacks. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### MAC Filtering
***

MAC filtering can also be implemented. 

And this uses a whitelist to allow only specified machines to access the network. 

It's also configured at the switch and does require a list of MAC addresses to be maintained. 

So there can be some administrative overhead associated with this. 

But if your environment is quite static, in other words, you aren't seeing a whole lot of change with respect to the systems that come and go. Then you certainly can configure these MAC addresses. And, in essence, anything that is not on the list will not be allowed to connect. 

That also can help to prevent the rogue DHCP server. If it's a system that is just brought in by someone attempting to circumvent your security. And that MAC address of that system, whatever it is, is not on the whitelist, then it will not be allowed to participate in network communications.

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### VLAN Assignment
***

And finally, you can implement VLAN assignment. 

And this involves segregating traffic into separate VLANs. 

So you can add another layer of security, as opposed to having everything within the same VLAN. You quite simply decide which types of resources maybe should be isolated based on some kind of category. So you can control access to your network resources independently of the underlying topology. Because we are, at the end of the day, all still plugged in into the same physical switches. But I can isolate my servers from my clients and/or perhaps particular services that they offer into separate VLANs. 

So, again, it's just another layer of protection. Rather than having everything exposed within the same VLAN, a little bit of isolation certainly will never hurt. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### Conclusion - Configurations at the Switch Level
***

So by implementing these different types of configurations at the switch level, you can improve the overall security of your environment.

###### 7. Network Segmentation
***

And all of those are different variations of access control lists, that which is acceptable and that which is not acceptable.

###### 12. Honeypots
***

After completing this topic, you should be able to: ``describe honeypots``

Now in this presentation, we'll take a look at Honeypots and Honeynets.

***
###### Decoy
***

By presenting what appears to be a legitimate but vulnerable system, but in fact, it's really just a decoy. 

And this can be very useful in terms of gathering information, because you can literally see what type of attack is being launched what the methods are, and maybe even where it's coming from. So this can help you to identify information about the intruder themselves. 

So the basic idea is, let's say you happen to be monitoring some network traffic, and you do actually catch some suspicious activity. 

Of course, you could entirely block the attack by taking that system offline unplugging it from the network for example. 

While that foils the attack, it doesn't allow you to gain any insight into the attack. So future attacks, of course, might still happen. So with the honeypot in essence you let them proceed with the attack because they are only operating against a decoy. So, you quite literally watch what happens. Now, this might just logged, of course, but the idea is that you actually see what they did, how they got in, what methods they are using so that you can better protect your environment in the future.

###### Firewall Design

Now you certainly need to consider what your firewall design is in terms of their placement.

Typically, systems such as Web servers need to be exposed to the internet, but you don't want them inside your corporate lab.

So you create this demilitarized zone which has a firewall facing the internet and another firewall facing the Internal network. And then those exposed systems in between, that's the demilitarized zone.

So quite often you would place the honey pots in the DMZ.

Whereas a honeyfarm is still a collection of honey pots but it's more so multiple individual targets not so much the network itself. 

But you might hear those two terms used some what interchangeably. But in either case it's just a more robust implementation of vulnerable systems as opposed to just a single vulnerable system.

###### 13. Penetration Testing
***

After completing this topic, you should be able to: ``recognize the benefits of penetration testing``

In this presentation, we'll talk about penetration testing or what is often just dubbed pen testing, for short.

But this is the process of actively, yet voluntarily, exploiting vulnerabilities with respect to the hardware and software you have implemented, the configurations you have. And even using social engineering techniques against your own users to, quite simply, find out what is the overall security of your organization. 

So, as mentioned, this is done voluntarily with the consent of system owners. 

Now in many cases, you might hire a third party to implement the testing, so you might have non-disclosure agreements in place. 

But they, in essence, launch these attacks against your environment so that they can report back and say here is where we were able to compromise and here is where we were blocked. 

So you, of course, can address the weaknesses. 

Now this can be done manually or through a variety of automated processes and depending on the size of your environment, can also take a fair amount of time. 

But in terms of the benefits, you get this overall picture as to how effective the security is in your organization. 

Now there might be multiple application testing phases, again, it can take a while. And you do need to be mindful of potential disruptions to your production systems. 

So they might launch some tests against just some maintenance or development systems first, just to see whether or not these vulnerabilities exist. But then they still have to ultimately try to compromise the production system. So in the event that they succeed, this, of course, could disrupt your environment. 

But ultimately, you need to know these things. 

< [Table of Contents](#top) | [References](#references) >

###### Black Box
***

Now, there are a few different implementations, but a black box penetration test is, in essence, where the attackers themselves, the testers, don't really know anything about the target. 

So they're kind of going in blind, if you will, hence the term black box. 

But in many cases, that is the perspective of an attacker. They really don't know what it is that they might find, but they're looking for anything. 

< [Table of Contents](#top) | [References](#references) >
<a id="___"></a>
***
###### White Box
***

A white box is the exact opposite, whereby the testers have details about the target. 

So they are launching much more focused attacks against that target to find out if, of course, they can compromise it. 

< [Table of Contents](#top) | [References](#references) >

###### Conclusion - Penetration Testing
***

And, in many cases, penetration testing might be implemented as a combination of both black box and white box.

So some examples, of course, could include impersonating a help desk technician and calling an accounting clerk. So that would be a social engineering technique. Just flat out finding out if I can get this person to divulge sensitive information. Maybe it's trying to break into a Wi-Fi network using WPA Personal encryption. Maybe it's simply reviewing application code prior to its deployment. Or seeing if I can successfully crack user passwords. And, again, there can be a variety of techniques that might be used in that case but, ultimately, it's just different types of attacks. And reporting on what the results were, so that you can go back and say here is what we were able to do. We circumvented your security here, here, and here. We were not successful here, here, and here. So you now know where the weakness are and, of course, you can address those. 

So penetration testing is very useful for tightening things up. And it's something that should also be done regularly. You shouldn't assume to configure everything one time then test it and basically go about your business from that point on thinking you will always be secure. It's something that you always need to keep on top of. And as things change and evolve, you should implement tests to keep up with those changes.
